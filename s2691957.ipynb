## Task 3: Spatial Lag & Moran's Plot

In Lab 4 (http://darribas.org/gds16/labs/Lab_04.html) by Dani Arribas, you can see examples of weight matrices and at the end an explanation of spatial lag and the Morans plot (Watch out, this is not Moran's I/ Moran's spatial autocorrelation, which is also often calculated in spatial analytics). 
Read and understand the concept of spatial lag first. In a Moran's plot, you can show how the value of a variable relates to it's spatial lag (or the average of this value in the surrounding areas). 

Additional information: 
* [Dani Arribas written course material](http://darribas.org/gds_scipy16/)
* [installing](https://pysal.org/install) and [getting started](https://pysal.org/getting_started) with PySAL.

```python
#Spatial lag, Morans Plot
knn4 = ps.weights.KNN.from_shapefile(africa_shp, k=4)
# Extract centroids
africa = africa.to_crs(epsg=3395)   #re-project geometries
cents = africa.centroid
pts = np.array([(pt.x, pt.y) for pt in cents])
knn4_from_pts = ps.weights.KNN.from_array(pts, k=4)
#Storing weight matrix in file
fo = ps.io.open(r'C:\Users\NAVEEN\Desktop\UC A2\Data\africa_knn.gal', 'w')
fo.write(knn4_from_pts)
fo.close()
```

```python
africa_g4 = africa.join(d4)   #For goal 4
africa_g6 = africa.join(d6)    #For goal 6
africa_g4.head()
```

```python
# Standardize the values (for goal 4)
std_g4 = (africa_g4['2018'] - africa_g4['2018'].mean()) / africa_g4['2018'].std()
std_w_g4 = pd.Series(ps.weights.lag_spatial(knn4_from_pts, std_g4), index=std_g4.index)  #Spatial lag for standardized data

# Setup the figure and axis
f, ax = plt.subplots(1, figsize=(9, 9))
sns.regplot(x=std_g4, y=std_w_g4)
ax.set_xlabel("Literacy rate, adult total (% of people ages 15 and above)")
ax.set_ylabel("Spatial Lag")
plt.title("Moran's plot for Goal 4")
plt.axvline(0, c='k', alpha=0.5)
plt.axhline(0, c='k', alpha=0.5)
plt.show()
```

```python
#Now we'll check in which quadrant the spatial lag for Rwanda lies
std_w_g4.loc['Rwanda']
```

```python
for i in range(len(countries)):
    if countries[i] == 'Rwanda':
        neighbor_indexes = list(knn4_from_pts[i])     #neighbours of Rwanda
for index in neighbor_indexes:
    print(countries[index], std_w_g4.loc[countries[index]])    #spatial lag of neighbours
```

***Choropleth Maps***

```python
# Choropleth Maps
import mapclassify as mc

goal4_lag = ps.weights.lag_spatial(knn4_from_pts, africa_g4['2018'])      #Spatial lag for goal 4
goal4_lag_Q10 = mc.EqualInterval(goal4_lag, k=10)
f, ax = plt.subplots(1, figsize=(9, 9))
africa.assign(cl=goal4_lag_Q10.yb).plot(column='cl', categorical=True, 
              k=10, cmap='OrRd', linewidth=0.1, ax=ax,                     #Orange-Red color scheme (shades of Orange and Red)
              edgecolor='black', legend=True)
ax.set_axis_off()
plt.title("Goal 4 Spatial Lag Deciles")
plt.show()
```

***Interpretation:***

The map corresponds to the findings in the Moran's plot. Most nations which are in proximity to each other have similar literacy
values. There are a few contrasting regions as well.
